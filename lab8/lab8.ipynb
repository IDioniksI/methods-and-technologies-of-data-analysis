```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest
```

```python
file_name = r'..\Dataset\adult_cleaned.data'
headers = ['Age', 'Work_class', 'Final_weight', 'Education', 'Education_num', 'Marital_status',
           'Occupation', 'Relationship', 'Race', 'Sex', 'Capital_gain', 'Capital_loss',
           'Hours_per_week', 'Native_country', 'Income']
df = pd.read_csv(file_name, sep=",", header=1, names=headers, na_values='?')
df
```

```python
print(df['Income'].unique())
```

Генеруємо випадкову вибірку з датасету

```python
random_df = df.sample(frac=0.4, random_state=42)
```

Розділяємо на групи

```python
group_A = df[df['Hours_per_week'] <= 30]
group_B = df[df['Hours_per_week'] > 30]
# Рандомна вибірка
random_group_A = random_df[random_df['Hours_per_week'] <= 30]
random_group_B = random_df[random_df['Hours_per_week'] > 30]
```

Обчислюємо метрики, щоб оцінити харктеристики кожної з груп та зробити порівняння

```python
mean_income_A = group_A['Income'].value_counts(normalize=True)['>50K']
mean_income_B = group_B['Income'].value_counts(normalize=True)['>50K']
avg_education_A = group_A['Education_num'].mean()
avg_education_B = group_B['Education_num'].mean()
# Рандомна вибірка
mean_income_random_A = random_group_A['Income'].value_counts(normalize=True)['>50K']
mean_income_random_B = random_group_B['Income'].value_counts(normalize=True)['>50K']
avg_education_random_A = random_group_A['Education_num'].mean()
avg_education_random_B = random_group_B['Education_num'].mean()
```

Робимо статистичний тест на пропорції

```python
count_main = [
    group_A['Income'].value_counts()['>50K'],
    group_B['Income'].value_counts()['>50K']
]
nobs_main = [len(group_A), len(group_B)]
z_stat_main, p_value_main = proportions_ztest(count_main, nobs_main)
# Рандомна вибірка
count_random = [
    random_group_A['Income'].value_counts()['>50K'],
    random_group_B['Income'].value_counts()['>50K']
]
nobs_random = [len(random_group_A), len(random_group_B)]
z_stat_random, p_value_random = proportions_ztest(count_random, nobs_random)
```

Виведення результатів:

```python
print("Метрики для основної вибірки:")
print(f"Частка доходу >50K у групі A: {mean_income_A:.2f}")
print(f"Частка доходу >50K у групі B: {mean_income_B:.2f}")
print(f"Середній рівень освіти у групі A: {avg_education_A:.2f}")
print(f"Середній рівень освіти у групі B: {avg_education_B:.2f}")

print("\nМетрики для рандомної вибірки:")
print(f"Частка доходу >50K у групі A: {mean_income_random_A:.2f}")
print(f"Частка доходу >50K у групі B: {mean_income_random_B:.2f}")
print(f"Середній рівень освіти у групі A: {avg_education_random_A:.2f}")
print(f"Середній рівень освіти у групі B: {avg_education_random_B:.2f}")

print("\nРезультати статистичного тесту для основної вибірки:")
print(f"z-статистика: {z_stat_main:.2f}")
print(f"p-значення: {p_value_main:.4f}")

print("\nРезультати статистичного тесту для рандомної вибірки:")
print(f"z-статистика: {z_stat_random:.2f}")
print(f"p-значення: {p_value_random:.4f}")

# Критерій успіху
alpha = 0.05
if p_value_main < alpha:
    print("\nВідхиляємо нульову гіпотезу для основної вибірки: частка доходу >50K у групах значущо різна.")
else:
    print("\nНе відхиляємо нульову гіпотезу для основної вибірки: частка доходу >50K у групах незначущо різна.")

if p_value_random < alpha:
    print("\nВідхиляємо нульову гіпотезу для рандомної вибірки: частка доходу >50K у групах значущо різна.")
else:
    print("\nНе відхиляємо нульову гіпотезу для рандомної вибірки: частка доходу >50K у групах незначущо різна.")
```
